# Part 2: Van der Waals Gas

```python
def model(P, V):
    n = 1
    R = 0.083144
    a = 16.02
    b = 0.1124
    T = (1 / (n * R)) * (V - n * b) * (P + a * ((n / V) ** 2))
    return T

pressure = np.linspace(1, 10, 10)
volume = np.linspace(10, 30, 20)
x, y = np.meshgrid(pressure, volume)
z = model(x, y)
# model(pressure, volume) doesn't sync size of p and v while meshgrid does

plt.figure()
plt.imshow(z, cmap='inferno', vmin=0, vmax=3500)
plt.title('Temperature')
plt.xlabel('Pressure')
plt.ylabel('Volume')
#for colorbar...
plt.colorbar()
plt.show()
```

# Part 3: Monte Carlo

```python
## 1e3
x = np.random.rand(1, 1000) # array of random numbers, automatically between 0 and 1
y = np.random.rand(1, 1000)

# finding distance from orgin
index = 0
inCircle = 0
while index < x.size:
    distance = ((x[0][index])**2 + (y[0][index])**2)**0.5
    if distance < 1:
        inCircle += 1
    index += 1
print('1e3 Pi Value: ' + str(4*inCircle/x.size))

## 1e4
x = np.random.rand(1, 10000) # array of random numbers, automatically between 0 and 1
y = np.random.rand(1, 10000)

# finding distance from orgin
plt.figure()
index = 0
inCircle = 0
while index < x.size:
    distance = ((x[0][index])**2 + (y[0][index])**2)**0.5
    if distance < 1:
        plt.scatter(x[0][index], y[0][index], s=4, color='red')
        inCircle += 1
    else:
        plt.scatter(x[0][index], y[0][index], s=4, color='blue')
    index += 1
print('1e4 Pi Value: ' + str(4*inCircle/x.size))
# Boundary Line
x2 = np.linspace(0, 1, 10000)
plt.plot(x2, (1 - x2**2)**0.5, color='orange')
plt.text(-5, 60, 'Pi Value: ' + str(4/x.size), fontsize = 22)
plt.show()

## 1e5
x = np.random.rand(1, 100000) # array of random numbers, automatically between 0 and 1
y = np.random.rand(1, 100000)

# finding distance from orgin
index = 0
inCircle = 0
while index < x.size:
    distance = ((x[0][index])**2 + (y[0][index])**2)**0.5
    if distance < 1:
        inCircle += 1
    index += 1
print('1e5 Pi Value: ' + str(4*inCircle/x.size))

## 1e6
x = np.random.rand(1, 1000000) # array of random numbers, automatically between 0 and 1
y = np.random.rand(1, 1000000)

# finding distance from orgin
index = 0
inCircle = 0
while index < x.size:
    distance = ((x[0][index])**2 + (y[0][index])**2)**0.5
    if distance < 1:
        inCircle += 1
    index += 1
print('1e6 Pi Value: ' + str(4*inCircle/x.size))
```
