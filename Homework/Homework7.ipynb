```python
import pandas as pd 
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

# Problem 1

```python
# loading in data
file = "AOC_recent_tweets.txt"
data = pd.read_json(file)
```

```python
# need to index tweets (id)
data.set_index('id', inplace=True)
```

```python
def time_in_hours(col):
    t = col.dt.hour + (col.dt.minute / 60) + (col.dt.second / (60 ** 2))
    return t
```

```python
# creating new hours column
data['hours'] = time_in_hours(data['created_at'])
# saving
newCol = ['created_at', 'hours', 'full_text']
newData = data[newCol]
newData
```

```python
newData.to_csv('Homework7Problem1.csv', index=False)
```

# Problem 2

```python
data = sns.load_dataset('planets')
data.dropna(subset=['year', 'method'], inplace=True) # remove null values
```

```python
plt.figure()
plt.grid()
plt.title('Orbital Period vs Planet Mass')
plt.xlabel('Orbital Period (days)')
plt.ylabel('Mass (Jupiter Mass)')

plt.xscale('log')
plt.yscale('log')

sns.scatterplot(x='orbital_period', y='mass', hue='method', data=data)

plt.show()
```

```python
plt.figure()
plt.grid()
plt.title('Exoplanets Discovered by Year')
plt.xlabel('Year')
plt.ylabel('Number of Planets')

sns.countplot(x='year', hue='method', data=data)

plt.show()
```
